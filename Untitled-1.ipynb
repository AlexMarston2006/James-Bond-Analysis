```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("james_bond_data_cleansed.csv")
print(data)
```

```python
import numpy as np
```

```python
x = np.array(data['imdb'])
y = np.array(data['rotten_tomatoes'])

plt.scatter(x, y)
plt.xlabel("IMDb rating")
plt.ylabel("Rotten Tomatoes")
plt.title("IMDb and Rotten Tomatoes")
plt.show()
```

```python
# Answer: There is a strong correlation between IMDb rating and rotten tomatoes as both have similar ratings for each movie
```

```python
x1 = np.array(data['imdb'])
y1 = np.array(data['bond_kills'])

plt.scatter(x1, y1)
plt.xlabel("IMDb Rating")
plt.ylabel("Bond Kills")
plt.title("IMDb Rating and Bond Kills")
plt.show()
```

```python
# Answer: There is no correlation between IMDb Rating and how many kills James Bond has
```

```python
x = (data["release_year"])
y = (data["movie_budget"])
y1 = (data["income_world"])

plt.plot(x, y, 'b-o', label = "Movie Budget")
plt.plot(x, y1, 'r-o', label = "Income World")
plt.xlabel("Year Released")
plt.legend()
plt.show()
```

```python
# Over time, as the budget increases, in general the world income rises with some exceptions
# The major exceptions are the 2015, and 2021 movies where despite the budget going up
# the income world actually went down from where it was
```

```python
year = (data["release_year"])
income = (data["income_world"])
budget = (data["movie_budget"])
ratio = (income / budget)

fig = plt.figure(figsize = (7,5))

plt.bar(year, ratio, color='blue', width= 0.5)
plt.xlabel("Release Year")
plt.ylabel("Income World-Budget Ratio")
plt.title("Income World-Budget ratio over time")
plt.show()
```

```python
# Based on this trend the producer for sure made less profit now than in the pass
```
